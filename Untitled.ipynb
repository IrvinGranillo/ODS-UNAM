```python
import pandas as pd
import plotly.express as px

# 1. Cargar los datos desde la carpeta 'data'
path = 'data/1.1.1.a_dc_1081_es.csv'
df = pd.read_csv(path)

# 2. Limpieza de datos
# Quitamos el total nacional (cvegeo 0) para que no afecte la escala del mapa
df_map = df[df['cvegeo'] != 0].copy()

# Convertimos la columna de pobreza a numérica
col_pobreza = 'Población_que_vive_por_debajo_del_umbral_internacional_de_la_pobreza'
df_map[col_pobreza] = pd.to_numeric(df_map[col_pobreza], errors='coerce')

# Diccionario para corregir nombres de estados y que coincidan con el GeoJSON
name_mapping = {
    'Coahuila de Zaragoza': 'Coahuila',
    'Michoacán de Ocampo': 'Michoacán',
    'Veracruz de Ignacio de la Llave': 'Veracruz',
    'México': 'Estado de México'
}
df_map['Estado'] = df_map['Entidad_federativa'].replace(name_mapping)

# 3. URL del GeoJSON de los estados de México
geojson_url = "https://raw.githubusercontent.com/angelnmara/geojson/master/mexicoHigh.json"

# 4. Crear el mapa interactivo
fig = px.choropleth(
    df_map,
    geojson=geojson_url,
    locations='Estado',          # Columna con el nombre del estado
    featureidkey='properties.name', # Clave interna del GeoJSON
    color=col_pobreza,           # Variable a representar
    animation_frame='Periodo',   # Slider por año
    color_continuous_scale="Blues", # <--- Color AZUL solicitado
    title='Población bajo el umbral de pobreza en México (2016-2020)',
    labels={col_pobreza: 'Personas (Miles)'},
    hover_name='Entidad_federativa'
)

# 5. Ajustar el ZOOM y estilo
fig.update_geos(
    fitbounds="locations", # <--- Ajusta el zoom automáticamente a los estados
    visible=False          # Oculta el resto del mundo para mejor enfoque
)

fig.update_layout(
    margin={"r":0,"t":50,"l":0,"b":0},
    coloraxis_colorbar=dict(title="Miles de personas")
)

# 6. Guardar y mostrar
fig.write_html('mapa_pobreza_mexico_azul.html')
fig.show()
```
