## METAS

````python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib.offsetbox import OffsetImage, AnnotationBbox

data = {
    "ODS": [
        "1. Fin de la pobreza", "2. Hambre cero", "3. Salud y bienestar", "4. Educación de calidad",
        "5. Igualdad de género", "6. Agua limpia", "7. Energía", "8. Trabajo decente",
        "9. Industria", "10. Desigualdades", "11. Ciudades", "12. Producción",
        "13. Acción clima", "14. Vida submarina", "15. Vida ecosistemas", "16. Paz y justicia", "17. Alianzas"
    ],
    "Metas": [9, 5, 11, 10, 10, 10, 6, 14, 9, 7, 9, 5, 4, 3, 5, 10, 9],
    "Color": [
        "#E5243B", "#DDA63A", "#4C9F38", "#C5192D", "#FF3A21", "#26BDE2", "#FCC30B", "#A21942",
        "#FD6925", "#DD1367", "#FD9D24", "#BF8B2E", "#3F7E44", "#0A97D9", "#56C02B", "#00689D", "#19486A"
    ],
    "Archivo": [
        "FIN_POBREZA.jpg", "CERO_HAMBRE.jpg", "SALUD_Y_BIENESTAR.jpg", "EDUCACION_DE_CALIDAD.jpg",
        "IGUALDAD_DE_GENERO.jpg", "AGUA_LIMPIA_Y_SANEAMIENTO.jpg", "ENERGIA_ASEQUIBLE.jpg", "TRABAJO_DECENTE.jpg",
        "INDUSTRIA.jpg", "REDUCCION_DE_DESIGUALDADES.jpg", "CIUDADES_Y_COMUNIDADES_SOSTENIBLES.jpg", "PRODUCCION_Y_CONSUMO_RESPONSABLE.jpg",
        "ACCION_POR_EL_CLIMA.jpg", "VIDA_SUBMARINA.jpg", "VIDA_ECOCISTEMAS_TERRESTRES.jpg", "PAZ_JUSTICIA.jpg", "ALIANZAS.jpg"
    ]
}

df = pd.DataFrame(data)
df['Ruta'] = 'imagenes/' + df['Archivo']

def limpiar_titulo(texto):
    partes = texto.split('. ', 1)
    return partes[1].upper() if len(partes) > 1 else texto.upper()

df['Etiqueta'] = df['ODS'].apply(limpiar_titulo)
df1 = df.iloc[:8].copy().reset_index(drop=True)
df2 = df.iloc[8:].copy().reset_index(drop=True)

FONDO = '#1B2631'
TEXTO = 'white'

def graficar_bloque(ax, dataframe):
    y_pos = range(len(dataframe))
    
    ax.set_facecolor(FONDO)
    
    ax.barh(y_pos, dataframe['Metas'], color=dataframe['Color'], height=0.95, edgecolor=FONDO, linewidth=0.5)
    
    for i, row in dataframe.iterrows():
        try:
            img = mpimg.imread(row['Ruta'])
            imagebox = OffsetImage(img, zoom=0.13) 
            
            ab = AnnotationBbox(imagebox, (0, i), frameon=False, 
                                xybox=(40, 0), boxcoords="offset points", pad=0)
            ax.add_artist(ab)
            
            texto_completo = f"{row['Etiqueta']}    |    {row['Metas']}"
            
            ax.text(2.0, i, texto_completo, color='white', fontsize=16, 
                    fontweight='bold', va='center', ha='left', fontfamily='sans-serif')
            
        except:
            pass

    ax.set_yticks([]) 
    ax.set_xlim(0, 15) 
    ax.invert_yaxis()
    
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    
    ax.spines['bottom'].set_color(TEXTO)
    ax.tick_params(axis='x', colors=TEXTO, labelsize=12)
    
    ax.grid(axis='x', linestyle=':', alpha=0.2, color='white')
    ax.set_xlabel('CANTIDAD DE METAS', fontsize=14, fontweight='bold', color=TEXTO)

fig, axes = plt.subplots(1, 2, figsize=(18, 10))
fig.patch.set_facecolor(FONDO)

graficar_bloque(axes[0], df1)
graficar_bloque(axes[1], df2)

plt.tight_layout()
plt.show()
````
