````python

import pandas as pd
import plotly.express as px

# 1. Datos del proyecto ODS
data = {
    "ODS": [
        "1. Fin de la pobreza", "2. Hambre cero", "3. Salud y bienestar", 
        "4. Educación de calidad", "5. Igualdad de género", "6. Agua limpia",
        "7. Energía", "8. Trabajo decente", "9. Industria", "10. Desigualdades",
        "11. Ciudades", "12. Producción", "13. Acción clima", "14. Vida submarina",
        "15. Vida ecosistemas", "16. Paz y justicia", "17. Alianzas"
    ],
    "Metas_Totales": [9, 5, 11, 10, 10, 10, 6, 14, 9, 7, 9, 5, 4, 3, 5, 10, 9],
    "Color": [
        "#E5233D", "#DDA73A", "#4CA146", "#C7212F", "#EF402D", "#27BFE6",
        "#FBC412", "#A31C44", "#F26A2E", "#E01483", "#F89D2A", "#BF8D2C",
        "#407F46", "#1F97D4", "#59BA47", "#136A9F", "#14496B"
    ]
}

df_resumen = pd.DataFrame(data)

# 2. Generar estructura de bloques sin etiquetas internas
rows = []
for _, row in df_resumen.iterrows():
    for _ in range(row['Metas_Totales']):
        rows.append({
            "ODS": row['ODS'],
            "Valor": 1,
            "Color": row['Color']
        })

df_plot = pd.DataFrame(rows)

# 3. Crear la gráfica (se eliminan hover_name y hover_data)
fig = px.bar(
    df_plot, 
    x="ODS", 
    y="Valor", 
    color="ODS",
    color_discrete_map=dict(zip(df_resumen['ODS'], df_resumen['Color'])),
    category_orders={"ODS": data["ODS"]}
)

# 4. Configuración estética: Líneas blancas y desactivación de hover
fig.update_traces(
    marker_line_color='white',
    marker_line_width=2,
    hoverinfo='skip' # <-- Esta línea elimina el texto "Meta X.Y" y cualquier tooltip
)

fig.update_layout(
    showlegend=False,
    xaxis_title=None,
    yaxis_title="Cantidad de Metas",
    plot_bgcolor='white',
    height=500,
    # Limpieza de ejes
    yaxis=dict(showgrid=False, zeroline=False),
    xaxis=dict(tickangle=-45)
)

fig.show()
````
