````python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


years = np.arange(2015, 2025)
poverty_rate = [43.2, 41.9, 41.0, 39.5, 41.9, 43.9, 38.0, 36.3, 35.5, 34.2] 

df = pd.DataFrame({'Año': years, 'Tasa de Pobreza (%)': poverty_rate})


plt.figure(figsize=(10, 6))


plt.plot(df['Año'], df['Tasa de Pobreza (%)'], marker='o', linestyle='-', color='#E5243B', linewidth=3, markersize=8, label='Pobreza Laboral')


plt.fill_between(df['Año'], df['Tasa de Pobreza (%)'], 30, color='#E5243B', alpha=0.1)


plt.title('Evolución Simulada de la Pobreza en México (2015-2024)', fontsize=16, fontweight='bold', color='#333')
plt.xlabel('Año', fontsize=12, fontweight='bold')
plt.ylabel('Porcentaje de Población (%)', fontsize=12, fontweight='bold')
plt.grid(True, linestyle='--', alpha=0.5)
plt.xticks(years)
plt.ylim(30, 46)


min_val = min(poverty_rate)
min_year = years[poverty_rate.index(min_val)]
plt.annotate(f'Mínimo: {min_val}%', xy=(min_year, min_val), xytext=(min_year-1, min_val-3),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.tight_layout()
````
